```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./Placement.csv')
```

```python
data.head()
```

```python
print(data.columns.values)
```

```python
print('='*50)
print('Describe data')
print('='*50)
print(data.describe())
```

```python
data = data.drop(['sl_no'],axis = 1)
```

```python
sns.countplot(data=data,x=data['status'])
```

```python
data['gender'].value_counts()
```

```python
graph = sns.countplot(data=data,x=data['gender'])
```

```python
df = pd.DataFrame(data.groupby(['gender','status'])['status'].count())
```

```python
df
```

```python
sns.countplot(data=data,x='gender',hue='status')
```

```python
#Conclusion : Males have high chances of getting placed compared to females
```

```python
#SSC Percentage
```

```python
sns.distplot(data['ssc_p'])
plt.title('Distribution of SSC Percentage')
plt.xlabel('SSC %')
```

```python
sns.catplot(y='ssc_p',x='status',data = data,hue='status')
plt.xlabel('Employment Status')
plt.ylabel('SSC %')
```

```python
data['ssc_b'].value_counts()
```

```python
df = pd.DataFrame(data.groupby(['ssc_b','status'])['status'].count())
```

```python
df
```

```python
sns.countplot(x='ssc_b',hue='status',data=data)
```
